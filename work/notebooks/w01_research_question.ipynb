## 1. My lane (or freestyle) and why

*Lane 4: CTR / Engagement Opportunity Scoring - I choose this lane because it focuses on identifying pages that receive search visibility but attract fewer clicks or lower user engagement than expected. The goal is to produce a ranked list of review candidates that helps content teams allocate their time more effectively. The recommendations support human decision-making rather than replacing it.*

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve?*

This project improves the decision of which content pages should be prioritized for CTR and engagement review. Since organizations often manage hundreds or thousands of pages, reviewing every page manually is impractical. The ranked recommendations help identify pages that receive meaningful search visibility but underperform in terms of clicks or user engagement, allowing teams to focus their efforts on the most promising opportunities.

*Who acts on it?*

The primary users are SEO specialists, content editors, and digital marketing teams. They use the ranked list to determine which pages should be reviewed first for improvements such as rewriting page titles, refining meta descriptions, improving search intent alignment, enhancing content quality, or monitoring future performance.

*What does a wrong recommendation cost?*

A false positive may cause the team to spend time reviewing a page that is already performing appropriately, reducing productivity and delaying work on higher-priority pages. A false negative is potentially more costly because a page with significant CTR or engagement improvement opportunities may be overlooked, resulting in continued loss of potential organic traffic and user engagement.

```python
con.sql(f"""
SELECT
COUNT(*) AS rows_with_impressions
FROM read_parquet('{rel}/fact_content_daily_performance/**/*.parquet')
WHERE gsc_impressions IS NOT NULL
""")
```

```python
con.sql(f"""
SELECT
AVG(gsc_impressions) AS avg_impressions,
AVG(gsc_clicks) AS avg_clicks,
AVG(gsc_clicks * 1.0 / NULLIF(gsc_impressions, 0)) AS avg_ctr
FROM read_parquet('{rel}/fact_content_daily_performance/**/*.parquet')
""")
```
